1.What is a random variable in probability theory?
--
In probability theory, a random variable is a mathematical function that assigns numerical values to the outcomes of a random experiment. It provides a way to quantify uncertainty by mapping each possible outcome to a real number.

2.What are the types of random variables?
--
Random variables in probability theory are classified into two main types based on the nature of their possible values:

1. Discrete Random Variables

A random variable is discrete if it can take on a countable number of distinct values. These values may be finite or countably infinite.

Characteristics:

The values are separated (not continuous).

Each value has a probability assigned to it.

Described using a probability mass function (PMF).

Examples:

The number of heads in 3 coin flips ({0, 1, 2, 3}).

The number of customers arriving at a store in an hour.

The roll of a six-sided die ({1, 2, 3, 4, 5, 6}).

2. Continuous Random Variables

A random variable is continuous if it can take on an infinite number of values within an interval of real numbers.

Characteristics:

The values are uncountable and form a continuous range.

Probabilities are assigned to intervals rather than specific values.

Described using a probability density function (PDF).

--
The Poisson distribution is a discrete probability distribution that models the number of times an event occurs in a fixed interval of time or space, assuming that the events occur independently and at a constant average rate.

It is particularly useful for modeling rare events that happen randomly but at an average rate, such as phone call arrivals, traffic accidents, or machine failures.

Number of patients arriving at an emergency room per hour.

Number of mutations in a DNA strand.

4. Business and Economics

Number of defective products in a batch of manufacturing.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(x, p):
    """
    Calculate the Probability Mass Function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): Probability of success (0 â‰¤ p â‰¤ 1).

    Returns:
    float: Probability P(X = x).
    """
    if x not in [0, 1]:  # Ensure x is a valid Bernoulli outcome
        return 0
    return (p ** x) * ((1 - p) ** (1 - x))

# Example Usage
p = 0.6  # Probability of success
print("P(X=0):", bernoulli_pmf(0, p))  # Probability of failure
print("P(X=1):", bernoulli_pmf(1, p))  # Probability of success
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_z_scores(data):
    """
    Compute Z-scores for a dataset.

    Parameters:
    data (array-like): Input numerical dataset.

    Returns:
    np.array: Z-scores of the dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation (n-1)
    z_scores = (data - mean) / std_dev
    return z_scores

# Generate a random normal dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Compute Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
sns.histplot(z_scores, bins=30, kde=True, color="skyblue", edgecolor="black")
plt.axvline(x=0, color="red", linestyle="--", label="Mean (Z=0)")

# Labels & Title
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.title("Histogram of Z-Scores")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
# 14.rite a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, data):
    """
    Compute the Z-score for a given data point.

    Parameters:
    x (float): The data point for which to compute the Z-score.
    data (array-like): The dataset to compute mean and standard deviation.

    Returns:
    float: Z-score of x.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation (n-1)
    z_score = (x - mean) / std_dev
    return z_score

# Generate a normal dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Example data point
x_value = 65

# Compute Z-score
z_value = calculate_z_score(x_value, data)

# Plot standard normal distribution for comparison
x = np.linspace(-4, 4, 1000)  # Standard Normal Range
pdf_values = norm.pdf(x, 0, 1)  # Standard Normal PDF

plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, 'b-', linewidth=2, label="Standard Normal PDF")

# Mark the Z-score location
plt.axvline(z_value, color="red", linestyle="--", label=f"Z-Score of {x_value} = {z_value:.2f}")

# Labels & Title
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.title("Z-Score Calculation and Standard Normal Distribution")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()

# Print result
print(f"The Z-score for x = {x_value} is: {z_value:.2f}")
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    sample (array-like): Sample data.
    population_mean (float): Hypothesized population mean (H0).
    population_std (float): Known population standard deviation.
    alpha (float): Significance level (default=0.05).

    Returns:
    tuple: Z-score, p-value, decision to reject or fail to reject H0.
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Compute Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Decision rule
    decision = "Reject H0" if p_value < alpha else "Fail to reject H0"

    return z_score, p_value, decision

# Example Dataset: Test if the average test score differs from 75
np.random.seed(42)
sample_data = np.random.normal(loc=77, scale=10, size=50)  # Sample from normal distribution
population_mean = 75  # Hypothesized mean
population_std = 10  # Known population standard deviation

# Perform Z-test
z_score, p_value, decision = z_test(sample_data, population_mean, population_std)

# Display Results
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Decision: {decision}")

```

```python
# 16.  Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Compute the confidence interval for the mean of a dataset.

    Parameters:
    data (array-like): The sample data.
    confidence (float): Confidence level (default=0.95 for 95% CI).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (n-1)

    # Z-critical value for the given confidence level
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Compute margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    # Compute confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

# Example Dataset: Test Scores
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=50)  # Mean=75, Std=10, n=50

# Compute 95% Confidence Interval
ci_lower, ci_upper, sample_mean = confidence_interval(sample_data)

# Display Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```

```python
# 19.  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define Poisson distribution parameter (lambda)
lambda_value = 5  # Average number of occurrences

# Generate x values (discrete values from 0 to 15)
x = np.arange(0, 16)

# Compute CDF values
cdf_values = poisson.cdf(x, lambda_value)

# Plot Poisson CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf_values, where='mid', color='blue', linewidth=2, label=f'Poisson CDF (Î»={lambda_value})')

# Highlight key points
plt.scatter(x, cdf_values, color='red', marker='o')

# Labels & Title
plt.xlabel("x (Number of Events)")
plt.ylabel("Cumulative Probability P(X â‰¤ x)")
plt.title("Cumulative Distribution Function (CDF) of a Poisson Distribution")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()

# Interpretation: Print a probability example
k_value = 7
prob = poisson.cdf(k_value, lambda_value)
print(f"P(X â‰¤ {k_value}) = {prob:.4f}, meaning there is a {prob*100:.2f}% chance that X is â‰¤ {k_value}.")

```

```python
# 23.  Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a normal distribution dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15

# Compute mean and standard deviation
mean = np.mean(data)
std = np.std(data, ddof=1)  # Sample standard deviation

# Apply Z-score normalization
normalized_data = (data - mean) / std

# Plot Histograms (Before & After Normalization)
plt.figure(figsize=(10, 5))

# Original Data
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color="skyblue")
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Normalized Data
plt.subplot(1, 2, 2)
sns.histplot(normalized_data, bins=30, kde=True, color="orange")
plt.axvline(0, color='red', linestyle='--', label="Mean = 0 (Standardized)")
plt.title("Z-Score Normalized Distribution")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Print summary
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std:.2f}")
print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Normalized Std Dev: {np.std(normalized_data, ddof=1):.2f}")

```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_distribution(size=1000, mean=50, std=15):
    # Generate a normal distribution dataset
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Compute skewness and kurtosis
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis (normal kurtosis = 0)

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (Excess): {kurtosis:.4f} (Normal = 0)")

    # Plot Histogram
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='skyblue')
    plt.axvline(np.mean(data), color='red', linestyle='--', label=f"Mean = {np.mean(data):.2f}")
    plt.title("Histogram of Normally Distributed Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# Run the function
analyze_distribution()
```
